```python
import pytesseract
from pdf2image import convert_from_path
from PIL import Image
import os

```

```python
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

PDF_PATH = r"../data/agreement.pdf"

POPPLER_PATH = r"C:\Users\rajal\Downloads\release-25.12.0-0\poppler-25.12.0\Library\bin"
```

```python
pages = convert_from_path(
    PDF_PATH,
    dpi=300,
    poppler_path=POPPLER_PATH
)

print("Total pages:", len(pages))
```

```python
full_text = ""

for i, page in enumerate(pages):
    text = pytesseract.image_to_string(page)
    full_text += f"\n--- Page {i+1} ---\n{text}"

print(full_text[:1000])  # preview first 1000 chars
```

```python
# Save extracted text to a file

OUTPUT_PATH = "../data/extracted_contract.txt"

with open(OUTPUT_PATH, "w", encoding="utf-8") as f:
    f.write(full_text)

print("✅ Extracted text saved to:", OUTPUT_PATH)
```
